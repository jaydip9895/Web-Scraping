**The CNN Moneyâ€™s Market Movers website (https://money.cnn.com/data/hotstocks/ ) tracks the most active stocks on a real time basis. Specifically, the most active, the top gainers and top losers are listed at any instance in time. You will first write Python scripts that collect the list of most actives, gainers and losers from the above website.** 


**Next, your programs should take the ticker symbols and names of these companies (and categories) and build a csv file (called stocks.csv) with data about each stock from the website: https://finance.yahoo.com/quote/AMD?p=AMD&.tsrc=fin-srch-v1 which gives the quote for ticker symbol AMD as an example. The data to be collected from the Yahoo Finance site should include:**


**OPEN price**

**PREV CLOSE price**

**VOLUME**

**MARKET CAP**



**Your code should also list the names of the companies in the order and categories listed in the website: https://money.cnn.com/data/hotstocks/ and ask the user to choose a company to get the data on. Once the user chooses the company of interest, your program should display its corresponding data (Open, Prev Close, Volume and Market Cap).**

```python
#import all useful libraries
import urllib.request
import urllib.parse
import urllib.error
from bs4 import BeautifulSoup, NavigableString, Tag
import pandas as pd
import ssl
import colorama
from colorama import Fore
import csv

#here we use the pandas library to get all the data from the website
#get data from the url and save it as pandas dataframe
url = 'https://money.cnn.com/data/hotstocks/'
df = pd.read_html(url, header=0)

print(Fore.BLACK+"\nThis is a program to scrape data from the {} for a class project.\n".format(url))
print(Fore.BLACK+"Which stock are you interested in:\n")

#print all company in the order and categories listed in the website
#according to most active, gainer and loser
print(Fore.BLACK+"Most Actives:")               #most active company list
for i in range(len(df[0])):
    print(Fore.BLACK+df[0]['Company'][i])
    
print(Fore.BLACK+"\nGainers:")                  #gainer company list
for i in range(len(df[1])):
    print(Fore.BLACK+df[1]['Company'][i])

print(Fore.BLACK+"\nLosers:")                   #looser company list
for i in range(len(df[2])):
    print(Fore.BLACK+df[2]['Company'][i])
    
#create list for company ticker name
#ticker name list for most active
ticker_list_1 = []                   
for i in range(len(df[0])):
    ticker1 = df[0]['Company'][i].split()[0]
    ticker_list_1.append(ticker1)

#ticker name list for gainer
ticker_list_2 = []
for i in range(len(df[1])):
    ticker2 = df[1]['Company'][i].split()[0]
    ticker_list_2.append(ticker2)

#ticker name list for looser
ticker_list_3 = []
for i in range(len(df[2])):
    ticker3 = df[2]['Company'][i].split()[0]
    ticker_list_3.append(ticker3)
    
#make final ticker list
#which contain all company ticke name: most active, gainer, loser 
#use extend as we want to add element in the list at lst
ticker_list = []
ticker_list.extend(ticker_list_1)    
ticker_list.extend(ticker_list_2)
ticker_list.extend(ticker_list_3)

#create list for company name 
#company name for most active stocks 
company_list_1 = []
for i in range(len(df[0])):
    company1 = df[0]['Company'][i].split()[1:]
    c_name_1 = " ".join(str(x) for x in company1)
    company_list_1.append(c_name_1)

#company name for most gainer stocks
company_list_2 = []
for i in range(len(df[1])):
    company2 = df[1]['Company'][i].split()[1:]
    c_name_2 = " ".join(str(x) for x in company2)
    company_list_2.append(c_name_2)

#company name for most loser stocks
company_list_3 = []
for i in range(len(df[2])):
    company3 = df[2]['Company'][i].split()[1:]
    c_name_3 = " ".join(str(x) for x in company3)               
    company_list_3.append(c_name_3)
    
#make final company list
#which contain all company ticke name: most active, gainer, loser 
#use extend as we want to add element in the list at lst
company_list = []
company_list.extend(company_list_1)
company_list.extend(company_list_2)
company_list.extend(company_list_3)

#make dictionary of ticker and company name
tic_com = dict(zip(ticker_list, company_list))

#define function for  catagories:most active, gainer, loser
#this is for make final list with catagories
def cat(i):
    if i<10:
        return 'Most Active'
    elif i>=10 and i<=19:
        return 'Gainer'
    else:
        return 'Loser'

#make list for catagories, company name and ticker name
tc_dict = []
for i in range(len(ticker_list)):
    #call the cat() function we defined erlier for print value for catagory key in dictionary
    initial = {'Category': cat(i), 'Ticker': ticker_list[i], 'Company Name': company_list[i]}
    tc_dict.append(initial)

#make list for differant values for company
#which are open, close, volumn, market cap
f_dict = []
for i in ticker_list:
    
    #for loop for run through all element of ticker name list 
    u = 'https://finance.yahoo.com/quote/'+i+'?p='+i+'&.tsrc=fin-srch-v1'
    data = pd.read_html(u, header=0)        #make dataframe of all data we get from website
    
    #make the data readable
    #remove first coloumn as it is data we need, not coloumn name 
    col_0 = []
    for c_0 in data[0].columns: 
        col_0.append(c_0)
        
    col_1 = []
    for c_1 in data[1].columns: 
        col_1.append(c_1)
        
    #give dataframe coloumn name 
    df0 = data[0].set_axis(['A', 'B'], axis=1, inplace=False)
    df1 = data[1].set_axis(['A', 'B'], axis=1, inplace=False)
    
    #add first coloumn data we removed at the end of the dataframe table
    df0.loc[7] = col_0
    df1.loc[7] = col_1
    
    #concat both dataframe and make one dataframe
    frames = [df0, df1]
    result = pd.concat(frames, ignore_index=True)
    
    #from dataframe make dictionary 
    mydict = dict(zip(result.A, result.B))

    #get useful value we need for our program and append to final dictionary
    val = {'Open':mydict.get('Open'),'Previous Close':mydict.get('Previous Close'),'Volume':mydict.get('Volume'),'Market Cap':mydict.get('Market Cap')}
    f_dict.append(val)

#list of [catagory,ticker,company]
list_1 = []
for i in range(30):
    qw_1 = list(tc_dict[i].values())
    list_1.append(qw_1) 

#list of differant value of company we needed
#[open,close,volumn,market cap]
list_2 = []
for i in range(30):
    qw_2 = list(f_dict[i].values())
    list_2.append(qw_2)

#list_1 became final list here which we gonna print in the .txt file
for i in range(len(list_1)):
    list_1[i].extend(list_2[i])

#this block is for if we want to save file as .txt file
#we will print list_1 in .txt file with seperator as ","
'''
with open("stocks.txt", "w") as f:
    print this line just to understand the format of the file
    f.write("Format of this file is:catagory,Ticker name,Company name,open,previous close,volumn,market cap\n")
    for row in list_1:
        f.write("%s\n" % ','.join(str(col) for col in row))
    f.close()
'''

#save data to .csv file as stocks.csv
#make dataframe of finam list which is list_1
my_df = pd.DataFrame(list_1)

#make list for give coloumn name to the csv file
Col = ['Catagory','Ticker Name','Company Name','Open','Previous Close','Volume','Market Cap']
#save all the result into csv file and give coloumn name 
my_df.to_csv('stocks.csv', index=False, header=Col)


#get input from user to print details
while True:
    
    #while loop for if user put wrong ticker name which is not in the ticker name list
    while True:
        
        #ask user ticker name 
        print(Fore.BLACK +"\nUser Inputs: {}".format(Fore.RED),end='')
        user_in = input()
        
        #make user input in lowercase and remove all space if user puts any by mistake
        u_in = user_in.upper().strip()
        
        #open the csv file 
        csv_data = pd.read_csv('stocks.csv')
        
        #check wether user input we have in the ticker list or not
        if u_in in list(csv_data['Ticker Name']):
            print(Fore.BLACK +"\nThe data for {0}, {1} is the following:\n".format(u_in, tic_com[u_in]))
            
            #get row of user input from dataframe 
            row = csv_data.loc[csv_data['Ticker Name'] == u_in]
            
            #get index number of that row from dataframe
            idx = row.loc[row['Ticker Name'] == u_in].index[0]
            
            #print the desired result using stored stocks.csv file
            print("{} {}".format(row['Ticker Name'][idx],row['Company Name'][idx]))
            print("Open: {}".format(row['Open'][idx]))
            print("Previous Close: {}".format(row['Previous Close'][idx]))
            print("Volume: {}".format(row['Volume'][idx]))
            print("Market Cap: {}".format(row['Market Cap'][idx]))
            
            break
        else:
            print(Fore.BLACK+"\nEnter Ticker Symbol from the List.")
    
    #ask user if they want to check agin or not
    print(Fore.BLACK +"\nDo you want to continue? [yes or no] {}".format(Fore.RED),end='')
    ans = input()
    if ans.lower() == 'no' or ans.lower() == 'n':
        break
    
print(Fore.BLACK+"\nGood Bye!!")
```
